```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('matches.csv')
df2 = pd.read_csv('deliveries.csv')
df.head()
```

```python
#1.  Count the total number of matches conducted in 2008.

len(df[df["season"] == 2008])
```

```python
#2.  Find the cities where the maximum and minimum number of matches were conducted. 

# print((df["city"].value_counts()).index[0])
# print((df["city"].value_counts()).index[-1])

df["city"].value_counts()


```

```python
#3. Find total count of matches city-wise.

df["city"].value_counts()
```

```python
#4. Tally the toss decisions each team has taken.

df.groupby("toss_winner")["toss_decision"].value_counts()
```

```python
#5. Count the total number of normal and tied matches.

df["result"].value_counts()
```

```python
#6. Find the teams where the result was a tie.

df[["team1", "team2", "result"]][df["result"] == "tie"]
```

```python
#7. Find the team which won the match by the highest and lowest number of runs.

print(df.loc[df["win_by_runs"].idxmax(), "winner"])
print(df.loc[df["win_by_runs"].idxmin(), "winner"])
```

```python
#8. Calculate mean, median and standard deviation of ‘win_by_runs’.

print(df["win_by_runs"].mean())
print(df["win_by_runs"].median())
print(df["win_by_runs"].std())
```

```python
#9. Find the venue where the team won by the highest and lowest number of runs.

print(df.loc[df["win_by_runs"].idxmax(), "venue"])
print(df.loc[df["win_by_runs"].idxmin(), "venue"])
```

```python
#10. Find the players who have won ‘Player of the Match’ more than 3 times.

for i in (df["player_of_match"].value_counts() >= 3).index:
    if (df["player_of_match"].value_counts() >= 3)[i]:
        print(i)
```

```python
#11. Find all deliveries where the batsman scored a six.

df2[df2["batsman_runs"] == 6]
```

```python
#12. Compute the average runs scored in matches in all the venues.

for i in set(df["city"]):
    count = [df["city"].value_counts()]
    print(count)
```

```python
#13. Find the umpires who umpired the maximum number of times.

df["umpire1"].value_counts()
```

```python
#19. Visualize toss decisions across all seasons.

df["toss_decision"].value_counts().plot()
```
